```python
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import model_from_json
json_file = open('modelexpression.json', 'r')
loaded_model_json = json_file.read()
json_file.close()
loaded_model = model_from_json(loaded_model_json)
# load weights into new model
loaded_model.load_weights("modelexpression.h5")
print("Loaded model from disk")
```

```python
def indexmax(a):
    return a.index(max(a))
```

```python
CATEGORIES = ["Angry", "Happy","Neutral","Sad","Surprise"]
vid=cv2.VideoCapture(0)
IMG_SIZE=224
while True:
    ret,frame=vid.read()
    faces = face_cascade.detectMultiScale(frame, 1.3, 5)
    for (x,y,w,h) in faces:
        frame = cv2.rectangle(frame,(x,y),(x+w+50,y+h+50),(255,0,0),2)
        region=frame[x:x+w+20,y:y+h+20]
        new_array = cv2.resize(region, (IMG_SIZE, IMG_SIZE))
        new_array=np.array(new_array).reshape(-1, IMG_SIZE, IMG_SIZE, 3)
        new_array=new_array/255.0
        print(loaded_model.predict(new_array))
        print(np.argmax(new_array))
        frame=cv2.putText(frame,CATEGORIES[np.argmax(loaded_model.predict(new_array))],(x,y),cv2.FONT_HERSHEY_SIMPLEX ,1,(0,255, 0),2,cv2.LINE_AA)
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
    cv2.imshow('frame',frame)
cv2.destroyAllWindows()      
```
